```python
#importing dependencies

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
from imblearn.over_sampling import SMOTE
```

```python
#loading dataset in pandas dataframe
crd_data= pd.read_csv('/content/creditcard.csv')
```

This Dataset is highly unbalanced

0- legit transaction
1- fraud transaction

```python
#seperating the Data for analysis
legit= crd_data[crd_data.Class==0]
fraud= crd_data[crd_data.Class==1]
print(legit.shape)
print(fraud.shape)
```

```python
X = crd_data.drop('Class', axis=1)
Y = crd_data['Class']

print(X)
print(Y)
```

```python
# splitting data into training and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)

print(X_train.shape)
print(X_test.shape)
```

```python
# Applying SMOTE to the training data
smote = SMOTE(sampling_strategy='minority', random_state=2)
X_train_resampled, Y_train_resampled = smote.fit_resample(X_train, Y_train)
```

```python
# Checking the new class distribution
print("Resampled class distribution:")
print(Y_train_resampled.value_counts())
```

Now our Dataset is balanced

Now Training Model

```python
# Training the Logistic Regression model
model = LogisticRegression(random_state=2,max_iter=1000)
model.fit(X_train_resampled, Y_train_resampled)
```

```python
Y_pred = model.predict(X_test)

# Evaluating the model
print("Confusion Matrix:")
print(confusion_matrix(Y_test, Y_pred))

print("Accuracy Score:")
print(accuracy_score(Y_test, Y_pred))
```

We have got very good accuracy and well result in confusion matrix

Now we can compare train data with test data

```python
Y_train_pred = model.predict(X_train)

# Calculating accuracy on training data
train_accuracy = accuracy_score(Y_train, Y_train_pred)
print("Training Accuracy:", train_accuracy)
```
